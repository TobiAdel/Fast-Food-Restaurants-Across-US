```python
threshold = 8 # Anything that occurs less than this will be removed.
value_counts = cleanone['name'].value_counts() # Entire DataFrame 
to_remove = value_counts[value_counts <=threshold].index
cleanone.replace(to_remove, np.nan, inplace=True)

#cleanone
cleanone['name'].value_counts()
```

```python
cleanone['Street Adress'] = cleanone['address']+ ', ' +cleanone['city']+', '+cleanone['province']+' ' +cleanone['postalCode']
ch_restaurant_df  = cleanone[["name","country","latitude","longitude","Street Adress","province"]]
ch_restaurant_df
```

```python
#ch_restaurant_df.to_csv('clean_restaurant.csv', index = False)
```

```python
ch_restaurant_df['price_level'] = ""
ch_restaurant_df['rating'] = ""
ch_restaurant_df['user_ratings_total'] = ""
ch_restaurant_df.head()
```

```python
for x, row in ch_restaurant_df.iterrows():
    street_address=str(ch_restaurant_df.at[x,"Street Adress"])
    rest_name=str(ch_restaurant_df.at[x,"name"])
    lat=str(ch_restaurant_df.at[x,"latitude"])
    lng=str(ch_restaurant_df.at[x,"longitude"])
    response=requests.get(f"https://maps.googleapis.com/maps/api/place/textsearch/json?query={street_address,rest_name}&key={gkey}&location={lat,lng}")
    answer=response.json()
    try:
        user_rating_t=(answer['results'][0]['user_ratings_total'])
    except(KeyError,IndexError):
        print(f"missing rating total {x}")
    try:
        user_rating=(answer['results'][0]['rating'])
    except(KeyError,IndexError):
        print(f"missing rating {x}")
    try:
        p_level=(answer['results'][0]['price_level'])
    except(KeyError,IndexError):
        print(f"missing pl {x}")
    
    ch_restaurant_df.at[x,'price_level'] = p_level
    ch_restaurant_df.at[x,'User_Rating_Total'] = user_rating_t
    ch_restaurant_df.at[x,'rating'] = user_rating
```

```python
#ch_restaurant_df.to_csv('fastfood_restaurant_review.csv', index = False)
ch_restaurant_df=ch_restaurant_df.drop(columns=['country','user_ratings_total'])
```

```python
ch_restaurant_df['rating'] = ch_restaurant_df['rating'].astype(float)
ch_restaurant_df.dtypes
```

```python
state_rating_average = ch_restaurant_df.groupby(['name','province'])['rating'].mean()
```

```python
overview = pd.DataFrame({
    "state rating average": state_rating_average,
    "state restaurant total": state_restaurant_total,
    })
```

```python
overview=overview.reset_index(drop=False)
```

```python
overview
```

```python
overview=overview.rename(columns={'name':'Restaurant'})
```

```python
overview.dtypes
```

```python
a= overview[overview["Restaurant"] == "McDonald's"]['state rating average'].idxmax() #HI
b=overview[overview["Restaurant"] == "McDonald's"]['state rating average'].idxmin() #NE
c=overview[overview["province"] == "HI"]['state rating average'].idxmin()
d=overview[overview["province"] == "NE"]['state rating average'].idxmax()
```

```python
print(a)
print(b)
print(c)
print(d)
```

```python
aa=overview.loc[(a)]
bb=overview.loc[(b)]
cc=overview.loc[(c)]
dd=overview.loc[(d)]
```

```python
aaa=pd.DataFrame(aa)
aaa=aaa.transpose()
bbb=pd.DataFrame(bb)
bbb=bbb.transpose()
ccc=pd.DataFrame(cc)
ccc=ccc.transpose()
ddd=pd.DataFrame(dd)
ddd=ddd.transpose()
```

```python
eee=aaa.append(bbb)
eee=eee.append(ccc)
eee=eee.append(ddd)
eee
```

```python
#double_state_bar=overview.plot(x="region", y=["Max", "Min"], kind="bar")
#double_state_bar.set_xlabel("States")
#double_state_bar.set_ylabel("Rating")
sra_plot=sns.catplot(x="province",y='state rating average', hue="Restaurant",data=eee,errorbar=None,kind='bar',height=4,aspect=1,palette=['#FFFF00',"#FFA500","#D30000"])
sra_plot.set_axis_labels("State","Average Rating")
sra_plot.set_xticklabels(['Hawaii','Nebraska'])
sra_plot.set(ylim=(0,5))
sra_plot.fig.suptitle("Extreme Ratings")
sra_plot.savefig("Extreme Rating.png")
```

```python
srt_plot=sns.catplot(x="province",y='state restaurant total', hue="Restaurant",data=eee,errorbar=None,kind='bar',aspect=.8,palette=['#FFFF00',"#FFA500","#D30000"])
srt_plot.set_axis_labels("State","Restaurant Total")
srt_plot.set_xticklabels(['Hawaii','Nebraska'])
srt_plot.set(ylim=(0,6600))
srt_plot.fig.suptitle("Extreme Locations")
```

```python
overview_scatter=overview[overview['Restaurant'].isin(["McDonald's","Burger King","Chick-fil-A"])]
```

```python
extreme_scatter=sns.scatterplot(data=overview_scatter,x="state restaurant total",y="state rating average",hue="Restaurant",style="Restaurant",palette=["#FFA500","#D30000",'#FFFF00'],legend="auto",)
extreme_scatter.set_title("Extreme Scatter")
extreme_scatter.grid()
extreme_scatter.set(xlim=(0,7000))
extreme_scatter.set(ylim=(0,5))
extreme_scatter.figure.savefig("Extreme Scatter.png")
```

```python
extreme_regression=sns.lmplot(x='state restaurant total',y="state rating average",hue="Restaurant",data=overview_scatter,ci=None,palette=["#FFA500","#D30000",'#FFFF00'],fit_reg=True,legend_out=True)
plt.title("Extreme Scatter")
extreme_regression.figure.savefig("Extreme Regression.png")
```
